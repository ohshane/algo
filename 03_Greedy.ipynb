```python
# 그리디

# [1] 당장 좋은 것만 선택하는 그리디
# [예제 3-1] 거스름디

def solution(n):
    coin_types = [500, 100, 50, 10]
    coin_counts = dict.fromkeys(coin_types, 0)
    for coin in coin_types:
        coin_counts[coin] = n // coin
        n %= coin
    return coin_counts

solution(1260)

# 만약에 화폐단위가 순차적으로 배수단위가 아니라면 (예: 500, 400, 100)
# 그리디 알고리즘을 사용할 수 없다.
```

```python
# [2] [실전문제] 큰수의 법칙

def solution():
    n, m, k = map(int, input().split())
    data = list(map(int, input().split()))

    data.sort(reverse=True)
    if data[0] == data[1]:
        return data[0] * m
    else:
        return (data[0]*k + data[1])*(m//(k+1)) + data[0]*(m%(k+1))
    
solution()
```
