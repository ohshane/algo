```python
# [4-1]

def solution():
    n = int(input())
    seq = input().split()

    location = [1, 1]
    move = {
        'R': ( 0,  1),
        'L': ( 0, -1),
        'U': (-1,  1),
        'D': ( 1,  1)
    }

    for s in seq:
        temp = location.copy()
        temp[0] += move[s][0]
        temp[1] += move[s][1]
        if 0 <= temp[0] <= n and 0 <= temp[1] <= n:
            location = temp

    return location

solution()
        
```

```python
# [4-2]
def solution():
    n = int(input())
    count = 0
    for i in range(n+1):
        for j in range(60):
            for k in range(60):
                if '3' in f'{i}{j}{k}':
                    count += 1
    return count

solution()
```

```python
# [2]
def solution():
    x, y = input()
    x = ord(x) - 96
    y = int(y)
    possible_locations = [
        [x-1, y-2],
        [x-1, y+2],
        [x+1, y-2],
        [x+1, y+2],
        [x-2, y-1],
        [x-2, y+1],
        [x+2, y-1],
        [x+2, y+1],
    ]
    count = 0
    for x, y in possible_locations:
        if x in range(1, 9) and y in range(1, 9):
            count += 1
    return count

solution()
```

```python
# [3]
def solution():
    n, m = map(int, input().split())
    x, y, position = map(int, input().split())
    location = [x, y]

    board = []
    for _ in range(n):
        board.append(list(map(int, input().split())))
    
    move = [
        [-1,  0],
        [ 0,  1],
        [ 1,  0],
        [ 0, -1],
    ]

    def apply_move(location, move):
        return [location[0]+move[0], location[1]+move[1]]

    count = 0
    no_move = 0
    explored = []
    while True:
        position -= 1
        if position < 0:
            position = 3

        temp = apply_move(location, move[position])
        if board[temp[0]][temp[1]] == 0 and temp not in explored:
            location = temp
            explored.append(temp)
            no_move = 0
            count += 1
        else:
            no_move += 1

        if no_move == 4:
            temp_position = position
            temp_position += 2
            temp_position %= 4
            temp = apply_move(location, move[temp_position])
            if board[temp[0]][temp[1]] == 0:
                location = temp
                no_move = 0
            else:
                break

    return count

solution()
```
