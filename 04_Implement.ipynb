```python
# [4-1]

def solution():
    n = int(input())
    seq = input().split()

    location = [1, 1]
    move = {
        'R': ( 0,  1),
        'L': ( 0, -1),
        'U': (-1,  1),
        'D': ( 1,  1)
    }

    for s in seq:
        temp = location.copy()
        temp[0] += move[s][0]
        temp[1] += move[s][1]
        if 0 <= temp[0] <= n and 0 <= temp[1] <= n:
            location = temp

    return location

solution()
        
```
