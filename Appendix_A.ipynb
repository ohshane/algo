```python
# list

data = [0,1,2,3]
print(data)

data.append(0)
print(data)

data.insert(1, 9)
print(data)

data.remove(0)
print(data)

data.reverse()              # no inplace
sorted(data, reverse=True)  # no inplace
data.sort(reverse=True)     # inplace
print(data)
```

```python
# itertools

from itertools import permutations
from itertools import combinations
from itertools import combinations_with_replacement
from itertools import product

data = ['A','B','C']
print(list(permutations(data, 3)))
print()

print(list(combinations(data, 2)))
print()

print(list(combinations_with_replacement(data, 2)))
print()

print(list(product(data, repeat=2)))
print()
```

```python
# itertools
# infinite iterators

from itertools import count, cycle, repeat
# count(start, [step])
# cycle(p)
# repeat(elem, [n])

from itertools import accumulate
data = [1,1,1,1,1]
print('accumulate()')
print(list(accumulate(data)))

from itertools import chain
print('chain()')
print(list(chain([1,2,3], [100,200,300])))
print(list(chain([1,2,3], 'ABC')))

print('chain.from_iterable()')
print(list(chain.from_iterable(
    [[1,2,3],
     [4,5,6],
     [7,8,9]]
)))

from itertools import groupby
print('groupby()')
import random
data = [f'{random.randint(0,9):0>3}_{chr(random.randint(65,67))}' for _ in range(10)]
for k, group in groupby(data, key=lambda x: x[4]):
    print(k, list(group))
```

```python
# heapq (min heap)

# -(value) for max heap
# data[0] is the smallest value

import heapq

data = [1,5,2,4,3]
heapq.heapify(data)
for i in range(1, 6):
    heapq.heappush(data, -i)
for i in range(len(data)):
    print(heapq.heappop(data), end=' ')
```

```python
# bisect
# sorted list

from bisect import bisect_left, bisect_right

data = [1,2,3,3,3,3,3,4]
data.sort()
print(bisect_left(data, 3))
print(bisect_right(data, 3))
```

```python
# deque (for stack/que)
# indexing/slicing are not implemented

from collections import deque

data = deque([1,2,3,4,5])
data.append(6)
data.appendleft(0)
print(data)

data.pop()
print(data)

data.popleft()
print(data)
```
